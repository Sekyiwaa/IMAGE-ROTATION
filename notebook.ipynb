```
from PIL import Image
img = Image.open('PROJECT PURPLE.jpg')
img

```

```
import math
import numpy as np
I_ARRAY=np.array(img)
#I_ARRAY
angle=int(input('Enter the value of your angle'))
```

```
angle = math.radians(angle)
angle
```

```
cosine = math. cos(angle)
print(cosine)
sine=math.sin(angle)
print(sine)
height = I_ARRAY.shape[0]
print(height)
width= I_ARRAY.shape[1]
print(width)
```

```
NEW_HEIGHT= round(abs(I_ARRAY.shape[0]*cosine)+abs(I_ARRAY.shape[1]*sine))+1
NEW_HEIGHT
NEW_WIDTH= round(abs(I_ARRAY.shape[1]*cosine)+abs(I_ARRAY.shape[0]*sine))+1
NEW_WIDTH
OUTPUT= np.zeros((NEW_HEIGHT,NEW_WIDTH,I_ARRAY.shape[2]))
OUTPUT.shape
```

```
original_centre_height   = round(((I_ARRAY.shape[0]+1)/2)-1) 
original_centre_height
original_centre_width    = round(((I_ARRAY.shape[1]+1)/2)-1) 
original_centre_width
```

```
new_centre_height= round(((NEW_HEIGHT+1)/2)-1)  
new_centre_height
new_centre_width= round(((NEW_WIDTH+1)/2)-1)
new_centre_width
```

```
for i in range(height):
    for j in range(width):
        y=I_ARRAY.shape[0]-1-i-original_centre_height                   
        x=I_ARRAY.shape[1]-1-j-original_centre_width  
        new_y=round(-x*sine+y*cosine)
        new_x=round(x*cosine+y*sine)
```

```
new_y=new_centre_height-new_y
new_x=new_centre_width-new_x
if 0 <= new_x <NEW_WIDTH and 0 <= new_y < NEW_HEIGHT and new_x>=0 and new_y>=0:
            OUTPUT[new_y,new_x,:]=I_ARRAY[i,j,:]   
```

```
pil_img = Image.fromarray((OUTPUT).astype(np.uint8))
pil_img
```

```
print(OUTPUT)
```
